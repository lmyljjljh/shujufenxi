```python
import tushare as ts 
import pandas as pd 
from pandas import DataFrame,Series 
import numpy as np 
 
#获取某只股票的历史行情数据 
#code:字符串形式的股票代码 
df=ts.get_k_data(code='600519',start='2023-01-01',end='2023-02-01') #不知道股票数据从哪年开始，可以指定一个比较早的日期。 
df 
```

```python
df['date']=pd.to_datetime(df['date']) 
df
```

```python
#将date这一列作为源数据的行索引 
df.set_index('date',inplace=True) 
df.head() 
```

```python
df1 = df['open']
df1
```

```python
lista = list(df1)
lista
```

```python
listb = list(df1.index)
listb
```

```python
listc = list(df['close'])
listc
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams["axes.unicode_minus"] = False
lines = plt.plot(listb, lista,listb,listc)
# 设置 x 轴和 y 轴的标签
plt.xlabel("时间")
plt.ylabel("钱数/元")

# 设置x轴标签
plt.xticks(rotation=90) # 旋转90度
# 添加标题
plt.title("每天开盘与收盘价格")
# 添加图例
plt.legend(lines, ['开盘', '收盘'], shadow=True, fancybox=True)
# 显示网格
plt.grid(b=True, axis='y', linewidth=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams["axes.unicode_minus"] = False
x = range(len(listb))
plt.plot(x, lista, marker='o', label='开盘')
plt.plot(x, listc, marker='o', label='关盘')

for i,j,k in zip(lista,listc,x):
    plt.text(k-1, i  + 1, i, family='SimHei', fontsize=8, fontstyle='normal')
    plt.text(k-1, j  + 1.3, j, family='SimHei', fontsize=8, fontstyle='normal')
plt.xticks(x, listb, rotation ='vertical')
plt.xticks(rotation=90)
plt.title('开盘与收盘数据对比')
plt.legend()
plt.show()
```
