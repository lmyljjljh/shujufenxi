```python
#【综合练习】：在58同城搜索框输入内容后，爬取前5页数据（标题，房间，价格），写入excel文件
#计算价格的平均值与标准差

from selenium import webdriver
# from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
# from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

driver = webdriver.Chrome()       # 打开chrome浏览器
url='https://wh.58.com/house.shtml?PGTID=0d30000c-0009-e04e-105c-1fd380ee1459&ClickID=1'
driver.get(url)  # 注意，网址一定要带http

#用代码向受控chrome发送点击命令，实现网页自动跳转
search_btn = driver.find_element(By.XPATH,'//span[@class="text ui-search-text"]/input')
search_btn.send_keys('光谷\n')    # 将关键字填入搜索框进行搜索
time.sleep(2)


#     /html/body/div[6]/div[2]/ul/li[2]/div[2]/h2/a   #复制的标题的xpath
#     /html/body/div[6]/div[2]/ul/li[3]/div[2]/h2/a

# 将当前页面转为dom结构，并获取网页数据
dom=etree.HTML(driver.page_source)
# [i.strip() for i in dom.xpath('/html/body/div[6]/div[2]/ul/li/div[2]/h2/a/text()')]  
# 以上两句能够正确爬出数据，因此开始代入循环


title=[i.strip() for i in dom.xpath('/html/body/div[6]/div[2]/ul/li/div[2]/h2/a/text()')]
#     /html/body/div[6]/div[2]/ul/li[2]/div[2]/p[1]
#     /html/body/div[6]/div[2]/ul/li[4]/div[2]/p[1]
    
room=[i.strip() for i in dom.xpath('//p[@class="room"]/text()')]
    # info=[i.strip() for i in dom.xpath('//p[@class="infor"]/text()')]
price=[i.strip() for i in dom.xpath('//div[@class="money"]/b/text()')]
#     wait=WebDriverWait(driver, 10)
#     pages = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="pager_wrap"]/div/a[4]/span')),"Error！") #点击下一页
df=pd.DataFrame({'title':title,'room':room,'price':price})
df

# 以上程序段爬出一页，接下来循环爬出即可，但频繁访问会导致ip被禁止访问
```
